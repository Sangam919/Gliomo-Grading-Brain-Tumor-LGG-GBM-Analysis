```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler; from sklearn.model_selection import train_test_split; from sklearn.linear_model import Perceptron, LogisticRegression; from sklearn.neighbors import KNeighborsClassifier; from sklearn.tree import DecisionTreeClassifier; from sklearn.svm import SVC; from sklearn.metrics import accuracy_score, confusion_matrix


#  Load the Dataset
df = pd.read_csv('TCGA_GBM_LGG_Mutations_all.csv', na_values=' ?')


print("Missing values in each column:")
print(df.isnull().sum())

#  Data Preprocessing
df1 = df.copy()
cols = [
    'Grade', 'Project', 'Case_ID', 'Gender', 'Age_at_diagnosis', 'Primary_Diagnosis', 'Race',
    'IDH1', 'TP53', 'ATRX', 'PTEN', 'EGFR', 'CIC', 'MUC16', 'PIK3CA', 'NF1',
    'FUBP1', 'RB1', 'NOTCH1', 'BCOR', 'CSMD3', 'SMARCA4', 'GRIN2A', 'IDH2', 'FAT4', 'PDGFRA'
]

le = LabelEncoder()
for col in cols:
    df1[col] = le.fit_transform(df1[col].astype(str))

#  Separating Input and Target
target = df1['Grade']
input = df1.drop(columns=['Grade'])

#  Splitting into Train and Test
x_train, x_test, y_train, y_test = train_test_split(input, target, test_size=0.25)

#  Removing Any Remaining Non-Numeric Columns (if any)
non_numeric_cols = x_train.select_dtypes(include=['object']).columns
x_train = x_train.drop(non_numeric_cols, axis=1)
x_test = x_test.drop(non_numeric_cols, axis=1)

# Feature Scaling
sc = StandardScaler()
xtrain_sc = sc.fit_transform(x_train)
xtest_sc = sc.transform(x_test)

# Objective : Perceptron
p = Perceptron()
p.fit(xtrain_sc, y_train)
train_pred = p.predict(xtrain_sc)
test_pred = p.predict(xtest_sc)
print("\n\n\nðŸ“Œ Perceptron")
print("Training Accuracy:", accuracy_score(y_train, train_pred))
print("Testing Accuracy:", accuracy_score(y_test, test_pred))
print("Confusion Matrix:\n", confusion_matrix(y_train, train_pred))

# Objective : Logistic Regression
lr = LogisticRegression()
lr.fit(xtrain_sc, y_train)
lr_train_pred = lr.predict(xtrain_sc)
lr_test_pred = lr.predict(xtest_sc)
print("\n\n\nðŸ“Œ Logistic Regression")
print("Training Accuracy:", accuracy_score(y_train, lr_train_pred))
print("Testing Accuracy:", accuracy_score(y_test, lr_test_pred))
print("Confusion Matrix:\n", confusion_matrix(y_train, lr_train_pred))

#  Objective : K-Nearest Neighbors
knn = KNeighborsClassifier()
knn.fit(xtrain_sc, y_train)
knn_train_pred = knn.predict(xtrain_sc)
knn_test_pred = knn.predict(xtest_sc)
print("\nðŸ“Œ K-Nearest Neighbors")
print("Training Accuracy:", accuracy_score(y_train, knn_train_pred))
print("Testing Accuracy:", accuracy_score(y_test, knn_test_pred))
print("Confusion Matrix:\n", confusion_matrix(y_train, knn_train_pred))

# Objective : Decision Tree
dt = DecisionTreeClassifier(criterion='entropy', max_depth=6)
dt.fit(xtrain_sc, y_train)
dt_train_pred = dt.predict(xtrain_sc)
dt_test_pred = dt.predict(xtest_sc)
print("\nðŸ“Œ Decision Tree Classifier")
print("Training Accuracy:", accuracy_score(y_train, dt_train_pred))
print("Testing Accuracy:", accuracy_score(y_test, dt_test_pred))
print("Confusion Matrix:\n", confusion_matrix(y_train, dt_train_pred))

#  Objective : Support Vector Machine
svc = SVC()
svc.fit(xtrain_sc, y_train)
svc_train_pred = svc.predict(xtrain_sc)
svc_test_pred = svc.predict(xtest_sc)
print("\n\n\nðŸ“Œ Support Vector Machine (SVM)")
print("Training Accuracy:", accuracy_score(y_train, svc_train_pred))
print("Testing Accuracy:", accuracy_score(y_test, svc_test_pred))
print("Confusion Matrix:\n", confusion_matrix(y_train, svc_train_pred))



print("\n\n\n")

# Convert Age_at_diagnosis to numeric (assuming it's in days, as per earlier context)
df['Age_at_diagnosis'] = pd.to_numeric(df['Age_at_diagnosis'], errors='coerce') / 365  # Convert to years
# Verify columns exist and clean up data
print("Columns in DataFrame:", df.columns.tolist())
print("Missing values:\n", df[['Grade', 'Gender', 'Race', 'Age_at_diagnosis']].isna().sum())

# Objective : Count of GBM vs LGG
grade_counts = df['Grade'].value_counts()
plt.figure(figsize=(6, 4))
grade_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Distribution of Glioma Grades')
plt.xlabel('Grade (0 = LGG, 1 = GBM)')
plt.ylabel('Number of Patients')
plt.grid(True)
plt.show()

# Objective : Gender distribution
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6, 4))
gender_counts.plot(kind='bar', color=['lightgreen', 'lightcoral'])
plt.title('Gender Distribution')
plt.xlabel('Gender')  # Avoid hardcoding labels until we know encoding
plt.ylabel('Count')
plt.grid(True)
plt.show()



# Objective: Mutation distribution for IDH1
idh1_dist = df.groupby(['Grade', 'IDH1']).size().unstack(fill_value=0)
plt.figure(figsize=(6, 4))
idh1_dist.plot(kind='bar', stacked=True, color=['grey', 'blue'])
plt.title('IDH1 Mutation Distribution by Grade')
plt.xlabel('Grade')
plt.ylabel('Patient Count')
plt.grid(True)
plt.show()

# Objective : Pie chart of mutated vs not-mutated TP53
tp53_counts = df['TP53'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(tp53_counts, labels=['Not Mutated', 'Mutated'], autopct='%1.1f%%', colors=['lightgray', 'red'])
plt.title('TP53 Mutation Status')
plt.show()


# Objective : Race-wise grade distribution
race_grade = pd.crosstab(df['Race'], df['Grade'])
plt.figure(figsize=(8, 6))
race_grade.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Grade Distribution across Races')
plt.xlabel('Race')
plt.ylabel('Number of Patients')
plt.grid(True)
plt.tight_layout()
plt.show()

```
